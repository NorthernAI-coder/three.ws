```python
import vertexai
from vertexai.preview.vision_models import ImageGenerationModel
import os

# --- Configuration ---
GCP_PROJECT_ID = "aerial-vehicle-466722-p5"  # @param {type:"string"}
GCP_REGION = "us-central1"         # @param {type:"string"}

# Initialize Vertex AI
vertexai.init(project=GCP_PROJECT_ID, location=GCP_REGION)

model = ImageGenerationModel.from_pretrained("imagegeneration@006")

# Create a directory to store the generated images
output_dir = 'source_images'
os.makedirs(output_dir, exist_ok=True)

# Define the base prompt for image generation
prompt_template = "A photograph of the meme beach cookout at sunset, Pepe and Wojak characters grilling, volleyball game, POZ towels, no work allowed sign, keep exact style and lighting consistent. Camera angle: {angle}."

# List of camera angles to generate
camera_angles = [
    "eye-level, centered",
    "from a low angle, looking up",
    "from a high angle, looking down",
    "from the left side",
    "from the right side",
    "close-up on the grill",
    "wide shot of the whole scene",
    "from behind the volleyball net"
]

print(f"Generating {len(camera_angles)} images...")

# Generate an image for each camera angle
for i, angle in enumerate(camera_angles):
    prompt = prompt_template.format(angle=angle)
    response = model.generate_images(
        prompt=prompt,
        number_of_images=1,
    )
    image = response.images[0]
    image.save(f"{output_dir}/view_{i+1:03d}.png")
    print(f"Saved view_{i+1:03d}.png")

print("Image generation complete.")
```
